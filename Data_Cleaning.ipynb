```python
def clean_rating(rating_str):
    try:
        return float(rating_str)
    except ValueError:
        return None
df['rating'] = df['rating'].apply(clean_rating)

mean_ratings = df[df['rating'].notna()].groupby('country')['rating'].mean()

for country in mean_ratings.index:
    new_rating_mask = (df['rating'].isnull()) & (df['country'] == country)
    df.loc[new_rating_mask, 'rating'] = mean_ratings[country]
```

```python
import re
def extract_hour_convert_to_24hr(time_str):
    if isinstance(time_str, str) and time_str.strip().lower() == 'flexible':
        return 'Flexible'    
    if isinstance(time_str, str):
        match = re.search(r'(\d{1,2})', time_str)
        if match:
            hour = int(match.group(1))
            if hour < 12 and 'pm' in time_str.lower():
                hour += 12
            return f'{hour:02}:00'    
    if isinstance(time_str, str):
        try:
            hour = int(time_str)
            return f'{hour:02}:00'
        except ValueError:
            pass
    return None
df['checkin_num'] = df['checkin'].apply(extract_hour_convert_to_24hr)
```

```python
def convert_to_numeric(time_str):
    time_mapping = {
        '01:00': 1,
        '06:00': 6,
        '08:00': 8,
        '09:00': 9,
        '10:00': 10,
        '11:00': 11,
        '12:00': 12,
        '13:00': 13,
        '14:00': 14,
        '15:00': 15,
        '16:00': 16,
        '17:00': 17,
        '18:00': 18,
        '19:00': 19,
        '20:00': 20,
        '21:00': 21,
        '22:00': 22,
        '23:00': 23,
        'Flexible': 'Flexible' 
    }
    
    return time_mapping.get(time_str, None)

df['checkin_final'] = df['checkin_num'].apply(convert_to_numeric)
```

```python
df['checkin_final'].value_counts()
```

```python
def extract_hour(time_str):
    try:
        parts = str(time_str).split()
        hour = int(parts[0])
        period = parts[2].lower()    
        if period == 'pm' and hour != 12:
            hour += 12
        elif period == 'am' and hour == 12:
            hour = 0
        
        return hour
    except (IndexError, ValueError):
        return None  
df['checkout_final'] = df['checkout'].apply(extract_hour)
```
